```python
import cv2
import os
```

```python
Datos = './Data/amarillo'
if not os.path.exists(Datos):
    print('Carpeta creada: ',Datos)
    os.makedirs(Datos)
cap = cv2.VideoCapture('./Test/Amarillo.mp4')

count = 1
grados = 0
escala = 1.0
ancho, alto = 200, 200 #Tamaño de imagen requerido
while True:
    #Leer frame
    ret, frame = cap.read()
    if ret:
        #Mientras haya frames que extraer lo hara (264 frames por categoria)
        if count <= 1000:
            #Obtener tamaño de imagen original
            (h, w) = frame.shape[:2]
            centro = (w / 2, h / 2)
            #Girar frame de 0 hasta 360 grados
            rotacion = cv2.getRotationMatrix2D(centro, grados, escala)
            imagenGirada = cv2.warpAffine(frame, rotacion, (w, h))
            #Recortar frame girado
            filaInicial = int(h * .25) 
            columnaInicial = int(w * .25) 
            filaFinal = int(h * .75)
            columnaFinal = int(w * .75)
            imagenRecortada = imagenGirada[filaInicial:filaFinal, columnaInicial:columnaFinal]
            #Redimencionar frame recortado a resolucion requerida
            frame_nuevo = cv2.resize(imagenRecortada, (ancho, alto))
            #Guardar frame en ruta indicada
            cv2.imwrite(Datos+'/objeto_{}.jpg'.format(count),frame_nuevo)
            print('Imagen guardada:'+'/objeto_{}.jpg'.format(count))
            count = count + 1
            grados = grados + 1
        if grados == 361:
            grados = 0 
        if count == 1001:
            break  
cap.release()
cv2.destroyAllWindows()
```
